# Swiggy Data analysis

Step-1 Import libraries 

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_excel("swiggydata.xlsx")
```

```python
print(df)
```

```python
df
```

```python
# to show rating of data
```

```python
def handleRating(value):
    value = str(value).split()
    value = value[0];
    return float(value)

df['Rating']= df['Rating'].apply(handleRating)
print(df.head())
```

```python
df.info()
```

```python
grouped_data = df.groupby("Restaurent Name")["vote"].sum()
result = grouped_data
plt.plot(result, c="brown", marker="o")
plt.xlabel("Type of restaurent", c="green", size=20)
plt.ylabel("votes", c="red", size=20)
```
